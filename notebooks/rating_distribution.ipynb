```python
import pandas as pd
df = pd.read_csv("../data/uber_reviews_sampled.csv")
print(df.head())
print(df.dtypes)
rating_counts = df["rating"].value_counts().sort_index()
rating_percent = df["rating"].value_counts(normalize=True).sort_index() * 100
rating_dist = pd.DataFrame({
    "count": rating_counts,
    "percentage": rating_percent.round(2)
})
print(rating_dist)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.hist(df["rating"], bins=range(int(df["rating"].min()), int(df["rating"].max()) + 2), align="left")
plt.xlabel("Rating")
plt.ylabel("Number of reviews")
plt.title("Rating distribution (sampled)")
plt.show()
```
