```python
# verify distribution of ratings before model training

# As we can see the the ratings are pretty much the same as the original dataset, 
# we can proceed with model training without any concerns about the distribution of ratings being altered during the tagging process.
import pandas as pd
df = pd.read_csv("../data/uber_reviews_taggedOriginal.csv")
print(df.head())
print(df.dtypes)
rating_counts = df["rating"].value_counts().sort_index()
rating_percent = df["rating"].value_counts(normalize=True).sort_index() * 100
rating_dist = pd.DataFrame({
    "count": rating_counts,
    "percentage": rating_percent.round(2)
})
print(rating_dist)
```
